```python
import pandas as pd

df= pd.read_csv("mamlaka_aid_distribution_dataset.csv")
print(df)
```

```python
df.head()
df.info()

df.sample()

#From the investication above I have noticed the dattype on the distribution_date 
# is string instead of date & time and the inconcistensy on the gender column.
```

```python
#changed the datatype on the distribution_date from string to date & time.
df["distribution_date"] = pd.to_datetime(df["distribution_date"])

# df.info()
```

```python
#clean the gender column
df["gender"] = df["gender"].str.upper()
df["gender"] = df["gender"].replace({
    "M": "MALE",
    "F": "FEMALE"
})
```

```python
# df.head(20)
```

```python
#check for the duplicated rows in the data
df.duplicated().sum()
```

```python
#drop the duplicates
df = df.drop_duplicates()
```

```python
#check for any missing values in the data
df.isnull().sum()
```

```python
#there are 50 missing ages which will be filled with the median ofthe age bracket

df["age"]= df["age"].fillna(df["age"].median())
```
